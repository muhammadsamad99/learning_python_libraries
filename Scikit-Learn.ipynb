```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = {'Hours_study': [2, 3, 4, 5, 6, 7, 8, 9, 10], 'Exam_score': [20, 30, 35, 45, 60, 70, 75, 90, 95]}

df = pd.DataFrame(data)

X = df[['Hours_study']]
Y = df[['Exam_score']]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = LinearRegression()

model.fit(X_train, Y_train)

user_input = float(input("Enter number of hours in decimal/integer, you study:"))

response = model.predict([[user_input]])[0]
print(f"Predicted Exam Score: {response[0]:.2f}")
```
