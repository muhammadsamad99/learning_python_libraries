```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country'] == 'Pakistan']
temp_df
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
temp_df = gap[gap['country'].isin(['Pakistan', 'India', 'China'])]
```

```python
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line', hue='country')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='smoker', row='sex')
```
