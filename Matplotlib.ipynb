```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Pie Chart

```python
areas = ['Marketing', 'Customer Support', 'HR', 'Development', 'Sales']
budget = [2, 4.3, 5, 1, 6]
```

```python
plt.pie(budget, labels=areas, autopct="%0.2f%%")
plt.show()
```

```python
x=[10, 20, 15, 40, 60, 59, 89]
y=[19, 9, 31, 87, 57, 63, 29]
plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary Comparison')
plt.show()
```

```python
year=[2020, 2021, 2022, 2023, 2024, 2025]
price=[45000, 50000, 60000, 39000, 23000, 75000]
plt.bar(year, price)
plt.show()
```
